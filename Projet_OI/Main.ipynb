### 2) Filtered data frame on exports and imports 

#### 2.1) Conversion of FCFP into EURO

```python
df_imp_exp = df[['Date','Valeur Imports','Poids Imports','DTI','Valeur Exports','Poids Exports','DTE']] #filtering columns
df_imp_exp.head()   
```

```python
df_test = df_imp_exp.iloc[0:-1,[1,3,4,6]].floordiv(120) #convert FCFP into Euro
df_test[['Date','Poids Imports','Poids Exports']] = df_imp_exp[['Date','Poids Imports','Poids Exports']]

first_column = df_test.pop('Date')
df_test.insert(0,'Date',first_column) #put date into the first position of the data frame

df_test.head()
```
